```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
import plotly.graph_objects as go
```

```python
DATA_DIR = "../input/usdbdt/"
```

```python
data_df = pd.read_csv(DATA_DIR+"USD-BDT-forex.csv")
data_df.head()
```

```python
fig = go.Figure(data=[go.Candlestick(x=data_df['Date'],
                open=data_df['Open'],
                high=data_df['High'],
                low=data_df['Low'],
                close=data_df['Close'])])

fig.show()
```

```python
fig = go.Figure(data=[go.Ohlc(x=data_df['Date'],
                open=data_df['Open'],
                high=data_df['High'],
                low=data_df['Low'],
                close=data_df['Close'])])

fig.show()
```
